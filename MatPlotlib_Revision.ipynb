# Divide the Page into 2 parts

```python
plt.figure()
plt.subplot(1,2,1) # 1 page divide 2 parts and make plot 1
plt.plot(x,np.sin(x))

plt.subplot(1,2,2) # for second part
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.figure()
plt.subplot(2,2,1) # 2 by 2 page is divived into 2by2 grid
plt.plot(x,np.sin(x))

plt.subplot(2,2,2) # for second part
plt.plot(x,np.cos(x))


plt.subplot(2,2,3) # for third part
plt.plot(x,np.tan(x))


plt.subplot(2,2,4) # for fourth part
plt.plot(x,np.cos(x+1))
plt.show()
```

# title of plot , name x-axis and y-axis

```python
plt.plot(x,np.sin(x-0),color='blue') 
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid()
plt.axis('equal')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('This is sin plot before real dataset execution to understand basic syntax')
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color='blue',label='Sinx') 
plt.plot(x,np.cos(x-0),color='red',label='Cosx')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.grid()
# plt.axis('equal')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('This is sin plot before real dataset execution to understand basic syntax')
plt.legend() # it will information of lebel insights
plt.show()
```

# Scatter Plot: why we need it
1. show the relationship b/w x and y

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(6,2)) 
# 6inches x axis, 2 inches y axis
plt.scatter(x,y,c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Scatter Plot")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='r',alpha=.5) # alpha contorl the intensity of color
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Scatter Plot")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='r',alpha=.9,marker='*')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Scatter Plot")
plt.show()
```

# Pie chart

```python
langs=['c','python','java']
students = [20,100,40]

plt.pie(students,labels=langs)
plt.title("Pie Chart")
plt.show()
```

```python
langs=['c','python','java']
students = [20,100,40]

plt.pie(students,labels=langs,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
my_explod = (0.0,1.0,0.0)
langs=['c','python','java']
students = [20,100,40]

plt.pie(students,labels=langs,autopct='%1.1f%%',explode=my_explod)
plt.title("Pie Chart")
plt.show()
```

```python
my_explod = (0.0,0.5,0.0)
langs=['c','python','java']
students = [20,100,40]

plt.pie(students,labels=langs,autopct='%1.1f%%',explode=my_explod,shadow=True)
plt.title("Pie Chart")
plt.show()
```

# 3D Plot in MatPlotLib

```python
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z,color='#FF5733',alpha=.5)
plt.xlabel('X Axis')
plt.ylabel("Y Axis")
# Z label by default will take
plt.show()
```
