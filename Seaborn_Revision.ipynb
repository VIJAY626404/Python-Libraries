# Catplot

# Joint Plot

```python
sns.jointplot(x=tips.total_bill,y = tips.tip)
```

```python
# oposite of total_bill--> Distribution of total_bill
# oposite of tip --> Distribution of tip
```

```python
sns.jointplot(x=tips.total_bill,y = tips.tip,data=tips,kind='reg')
# regression line--> reg
```

```python
sns.jointplot(x=tips.total_bill,y = tips.tip,data=tips,kind='hex')
```

```python
sns.jointplot(x=tips.total_bill,y = tips.tip,data=tips,kind='kde')
# contour , used in geophysics,metrological department
```

# Pairplot : 
1. it will plot all the pairwise relationship of the dataframe

```python
sns.pairplot(tips) # numerical data
```

```python
sns.pairplot(tips,hue='sex')  # categorical data
```

```python
sns.pairplot(tips,hue='smoker',palette='coolwarm')
```

# Box Plot:
1. box plot used to determine outliars
2. extrem data are outliars

```python
# 1,2,3,4,5
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# data beyond upper fence and lower fence is called outliars
# upper fence = Q3 + 1.5 x IQR, lower fence = Q1-1.5 x IQR
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(data=tips,palette='rainbow',orient = 'h')
```

# correlation-->
. it is  statistical relationhip between two variable using heatmap

```python
tips_num = tips[['total_bill','tip','size']]
```

```python
tips_num.corr()
```

```python
sns.heatmap(tips_num.corr(),annot=True)
```

```python
sns.heatmap(tips_num.corr(),annot=True,cmap='coolwarm')
```

# Pie Chart

```python
my_explod = (0.0,0.5,0.0)
langs=['c','python','java']
students = [20,100,40]

plt.pie(students,labels=langs,autopct='%1.1f%%',explode=my_explod,shadow=True)
plt.title("Pie Chart")
plt.show()
```
