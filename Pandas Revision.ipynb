```python
# Merge and Join
```

```python
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]})
```

```python
df2 = pd.DataFrame({'key1':[1,2,45,6,67],
                   'key4':[4,5,6,7,8],
                   'key5':[3,56,5,6,6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,'left')  # left merge
# """
# 1. df1-->left data frame, df2--. right data frame
# 2.all element of left data frame and common element from both
# 3. Non common element in right side will print NaN
# """
```

```python
pd.merge(df1,df2,'right')  # right merge
```

```python
pd.merge(df1,df2,'outer') 
# outer element of left data frame + outer element of rightdata frame
# all the values of both data frame
```

```python
pd.merge(df1,df2,'inner') # only common element 
```

```python
pd.merge(df1,df2,'cross')
"""
1. cartesian product
2. try to combine all cobmbination of data frame
"""
```

# what happaned if not have common key in dataframe

```python
pd.merge(df1,df2,how='left',left_on='key2',right_on= 'key4')
# we are passing specific common element
# here keys are rename because two data don't have common name
```
